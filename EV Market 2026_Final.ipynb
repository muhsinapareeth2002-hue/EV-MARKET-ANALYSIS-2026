<h3 style="color: brown;">Visualization</h3>

<p>Drive Type Distribution Using BAR CHART</p>

```python
drive_counts = df['drive_type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(
    drive_counts.index,
    drive_counts.values
)
plt.title("Drive Type Distribution")
plt.xlabel("drive Type")
plt.ylabel("count")
plt.show()
```

LINE CHART - Average EV Price by Year

```python
year_price = df.groupby('year')['price_usd'].mean()
plt.figure(figsize=(8,5))
plt.plot(
    year_price.index,
    year_price.values,
    marker='o'
)
plt.title("Average EV Price by Year")
plt.xlabel("Year")
plt.ylabel("Average Price (USD)")
plt.grid(True)
plt.show()
```

SCATTER CHART - Battery Capacity vs Range

```python
plt.figure(figsize=(7,5))
plt.scatter(
    df['battery_capacity_kwh'],
    df['range_miles']
)
plt.title("Battery Capacity vs Range")
plt.xlabel("Battery Capacity (kWh)")
plt.ylabel("Range (Miles)")
plt.show()
```

HISTOGRAM - EV Price Distribution

```python
plt.figure(figsize=(7,5))
plt.hist(
    df['price_usd'],
    bins=10
)
plt.title("EV Price Distribution")
plt.xlabel("Price (USD)")
plt.ylabel("Frequency")
plt.show()
```

PIECHART - EV Market Segment Distribution

```python
segment_counts = df['market_segment'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(
    segment_counts.values,
    labels=segment_counts.index,
    autopct='%1.1f%%'
)
plt.title("EV Market Segment Distribution")
plt.show()
```

BOX PLOT - Annual Sales by Body Type

```python
plt.figure(figsize=(10,6))
sns.boxplot(
    x='body_type',
    y='annual_sales_units',
    data=df
)
plt.title("Annual Sales by Body Type")
plt.xlabel("Body Type")
plt.ylabel("Annual Sales")
plt.xticks(rotation=45)
plt.show()
```

SUBPLOT 

```python
fig, axes = plt.subplots(2, 2, figsize=(10,8))


# 1. Brand Count
brand_counts = df['brand'].value_counts().head(5)
axes[0,0].bar(
    brand_counts.index,
    brand_counts.values
)
axes[0,0].set_title("Top 5 Brands")
axes[0,0].tick_params(axis='x', rotation=45)


# 2. Customer Rating Histogram
axes[0,1].hist(
    df['customer_rating'],
    bins=10
)
axes[0,1].set_title("Customer Rating Distribution")
axes[0,1].set_xlabel("Rating")


# 3. Annual Sales Scatter Plot
axes[1,0].scatter(
    df['customer_rating'],
    df['annual_sales_units']
)
axes[1,0].set_title("Customer Rating vs Annual Sales")
axes[1,0].set_xlabel("Customer Rating")
axes[1,0].set_ylabel("Annual Sales")


# 4. Body Type Pie Chart
body_counts = df['body_type'].value_counts()
axes[1,1].pie(
    body_counts.values,
    labels=body_counts.index,
    autopct='%1.1f%%'
)
axes[1,1].set_title("Body Type Distribution")


plt.tight_layout()

plt.show()
```

COUNT PLOT

```python
plt.figure(figsize=(8,5))
sns.countplot(
    x='body_type',
    data=df
)
plt.title("Body Type Distribution")
plt.xlabel("Body Type")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.show()
```
